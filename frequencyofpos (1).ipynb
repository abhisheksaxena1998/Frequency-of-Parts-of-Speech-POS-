```python
import csv
import pandas as pd
import spacy

nlp = spacy.load('en_core_web_sm')

df=pd.read_csv('finalFile-9may.csv')
df.columns=['A','B','T']        

#print (df.head())     

tweets=df['T']

import spacy

nlp = spacy.load('en_core_web_sm')
i=0
for tweet in tweets:
   
    count=0
    countadj=0
    countverb=0
    countadp=0
    countadv=0
    countnum=0
    doc = nlp(tweet)
    for token in doc:
        #with open ('m4.csv','a') as res:
            #writer=csv.writer(res)
            #writer.writerows({'text','noun','other'})
            if token.pos_=='NOUN':
                count+=1
            if token.pos_=='ADJ':
                countadj+=1
            if token.pos_=='VERB':
                countverb+=1
            if token.pos_=='ADP':
                countadp+=1
            if token.pos_=='ADV':
                countadv+=1
            if token.pos_=='NUM':
                countnum+=1
                
                #writer.writerow({token.text:'t',token.pos_:'p'})
                #writer.writerow({token.text:'t',count:'p'})
                #print (f"{token.text} {token.pos_}")
    i+=1            
    print (f"nouns in tweet at {i} index are {count} verbs are {countverb} adjectives are {countadj} adpositions are {countadp} adverbs are {countadv} numerals are {countnum}")
    with open ('m249.csv','a') as res:
            writer=csv.writer(res)
            writer.writerow({i:'index',count:'verbs',countverb:'countofverbs',countadj:'cadj',countadp:'cadp',countadv:'cadv',countnum:'cnum'})
            #someerrorwriter.writerow(f"['{i}','{count}','{countverb}','{countadj}','{countadp}','{countadv}','{countnum}']")
df=pd.read_csv('m249.csv')
df.columns=['Index','Nouns','Verbs','Adjectives','Adpositions','Adverbs','Numerals']
print (df.head())
```
